## how connected are neighourhoods to the transit system

```python
from collections import defaultdict
```

```python
nh_structure = {'Kitchener': defaultdict(dict),
                'Waterloo': defaultdict(dict),
                'Cambridge': defaultdict(dict)}
```

```python
for row in kwc_nh.itertuples():
    nh_name = row.name
    nh_id = row.ID
    nh_city = row.city
    #bus_stop_count = int(sum(row.geometry.contains(grt_stops['geometry'])))
    routes = defaultdict(list)
    stops = defaultdict(dict)
    for stop in grt_links.itertuples():
        stop_in_nh = row.geometry.contains(stop.geometry)
        if stop_in_nh:
            stops[stop.StopID][stop.Route] = stop.FullName
    for row2 in grt_routes.itertuples():
        bus_num = row2.Route
        bus_dir = row2.Direction
        bus_name = row2.FullName
        intersects = row2.geometry.intersects(row.geometry)
        if intersects:
            routes[bus_num].append((bus_num, bus_dir, bus_name))
    bus_stop_count = len(stops.keys())
    route_intersections = len(routes.keys())
    nh_structure[nh_city][nh_name] = {'stop_count' : bus_stop_count, 
                                      'intersections': route_intersections,
                                      'routes': routes,
                                      'stops': stops}
    
    
```

```python
nh_structure['Kitchener']['VICTORIA HILLS'] # how many bus stops and how many routes are in the neighbourhood
```

```python
nh_structure = None
with open('./data/nh_bus_data.json') as f:
    nh_structure = json.load(f)
```

```python
nh_structure['Kitchener']['VICTORIA HILLS']
```

```python
import pandas as pd
city = []
nh = []
sc = []
ri = []
for c in nh_structure.keys():
    for h in nh_structure[c].keys():
        city.append(c)
        nh.append(h)
        sc.append(int(nh_structure[c][h]['stop_count']))
        ri.append(int(nh_structure[c][h]['intersections']))
```

```python
stop_route_df = pd.DataFrame(data={'city': city, 'name': nh, 'stops': sc, 'route_count': ri})
```

```python
stop_route_df
```

```python
nh_data_new = kwc_nh.merge(stop_route_df[['city', 'name', 'stops', 'route_count']], on=['city', 'name'], how='inner')
```

## Visualize the neighbourhoods and routes

```python
def style(feature):
    import ast
    ''' 
    taken from:
    https://gis.stackexchange.com/questions/394219/folium-draw-polygons-with-distinct-colours
    leverages the fact that each city has a unique colour attribute in the feature space
    '''
    return ast.literal_eval(feature['properties']['colour'])
```

```python
style1 = {'fillColor': '#228B22', 'color': '#228B22'}
style2 = {'fillColor': '#FF0000', 'color': '#FF0000'}
colours = kwc_nh.colour
m = folium.Map([43.442, -80.514], zoom_start = 10)
folium.GeoJson(grt_routes, style_function = lambda x: style2).add_to(m)
folium.GeoJson(kwc_nh,
              style_function = style).add_to(m)
m
```
