```python
def points_to_geojson(dct):
    '''
    Take the dictionary of polygons and builds a geojson object
    before returning it as json
    '''
    geojson = {'type':'FeatureCollection', 'features':[]}
    for k in dct.keys():
        feature = {'type':'Feature',
                       'id': k,
                       'geometry':{'type':'Polygon',
                                   'coordinates':[]}}
        feature['geometry']['coordinates'].append(dct[k])
        #feature['properties']['name'] = k

        geojson['features'].append(feature)
    return json.dumps(geojson)
```
